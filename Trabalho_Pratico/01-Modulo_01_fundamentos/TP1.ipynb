```python
# leitura dos dados
enem = pd.read_csv(
    './enem2019/DADOS/MICRODADOS_ENEM_2019.csv',
    sep = ';',
    decimal = ',',
    encoding= 'ISO-8859-1'
)
```

```python
enem
```

```python
# TP_sexo
enem['TP_SEXO_NOVO'] = enem.TP_SEXO.replace({
    'M' : 'Masculino',
    'F' : 'Feminino'
})

# TP_ESTADO_CIVIL
# 0	Não informado
# 1	Solteiro(a)
# 2	Casado(a)/Mora com companheiro(a)
# 3	Divorciado(a)/Desquitado(a)/Separado(a)
# 4	Viúvo(a)
enem['TP_ESTADO_CIVIL_NOVO'] = enem.TP_ESTADO_CIVIL.replace({
    0 : 'Não informado',
    1 : 'Solteiro(a)',
    2 : 'Casado(a)/Mora com companheiro(a)',
    3 : 'Divorciado(a)/Desquitado(a)/Separado(a)',
    4 : 'Viúvo(a)'
})

# TP_COR_RACA
# 0	Não declarado
# 1	Branca
# 2	Preta
# 3	Parda
# 4	Amarela
# 5	Indígena 
enem['TP_COR_RACA_NOVO'] = enem.TP_COR_RACA.replace({
    0 : 'Não declarado',
    1 : 'Branca',
    2 : 'Preta',
    3 : 'Parda',
    4 : 'Amarela',
    5 : 'Indígena'
})

# TP_ANO_CONCLUIU ensino medio
# 0	Não informado
# 1	2018
# 2	2017
# 3	2016
# 4	2015
# 5	2014
# 6	2013
# 7	2012
# 8	2011
# 9	2010
# 10	2009
# 11	2008
# 12	2007
# 13	Antes de 2007
enem['TP_ANO_CONCLUIU_NOVO'] = enem.TP_ANO_CONCLUIU.replace({
    0 : 'Não informado',
    1 : '2018',
    2 : '2017',
    3 : '2016',
    4 : '2015',
    5 : '2014',
    6 : '2013',
    7 : '2012',
    8 : '2011',
    9 : '2010',
    10 : '2009',
    11 : '2008',
    12 : '2007',
    13 : 'Antes de 2007'
})
# TP_ESCOLA
# 1	Não Respondeu
# 2	Pública
# 3	Privada
# 4	Exterior
enem['TP_ESCOLA_NOVO'] = enem.TP_ESCOLA.replace({
    1 : 'Não Respondeu',
    2 : 'Pública',
    3 : 'Privada',
    4 : 'Exterior'
})
# TP_ENSINO
# 1	Ensino Regular
# 2	Educação Especial - Modalidade Substitutiva
# 3	Educação de Jovens e Adultos

enem['TP_ENSINO_NOVO'] = enem.TP_ENSINO.replace({
    1 : 'Ensino Regular',
    2 : 'Educação Especial - Modalidade Substitutiva',
    3 : 'Educação de Jovens e Adultos'
})
```
